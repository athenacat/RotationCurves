### Basic Scatter Plots

```python
# Define some fake data
# You can just replace this with your own relevant data
# These functions are totally meaningless

# x-values
x = np.linspace(0,50,26)

#y-values for the first sample
y0 = 5.+0.3*x+np.random.uniform(low=-3., high=3., size=(26,))

#y-values for the second sample
y1 = 0.005*x**2.+np.random.uniform(low=-2., high=2., size=(26,))

#y-values for the third sample
y2 = 20.*np.exp(-x/3.)+np.random.uniform(low=-3., high=3., size=(26,))
```

```python
# This sets the dimensions of the axes (from 0 to 1)
# Values indicate [xmin,ymin,xmax,ymax]
# Make the plot smaller than the window to avoid cutting off titles, labels, legends, etc.
a = plt.axes([0.18, 0.18, .64, .64])

# Grid lines make any plot prettier
# ls sets the linestyle, standard options ('-','--',':',etc.)
# I like dot-dash but that's preference
# alpha affects transparency [0,1]
plt.grid(True,ls='-.',alpha=.4)

# Set the title
# I usually use fontsize 18
# The '\n' adds a line break
# Begining with r'string' is for LATEX inputs
plt.title(r'Meaningless Plots:'+'\n'+'Scatter',fontsize=18)

# Set the axes labels and tick sizes
# I usually use 16 for labels and ticks 
# Try using LATEX for symbols and special characters
plt.xlabel(r'$\mathcal{X}$ values',fontsize=16)
plt.ylabel(r'$\mathcal{Y}$ values',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Set the limits on your axes
# If this is not set it defaults to show every data point
# Sometimes these aren't really needed
plt.ylim(-10,25)
plt.xlim(-3,53)

# Here is where you plot your data
# I've done several styles for you to choose from
# I don't like plt.scatter so I use plt.plot but set ls='', which is no line
# Markers can have different shapes ('s','o','>','*', and some more exotic ones)
# Some people like markers with colored edges, some don't. This shows both styles
# ms sets the marker size, mfc is marker face color, mec is marker edge color
# matplotlib has tons of colors, see https://matplotlib.org/examples/color/named_colors.html
# label is a string to feed to plt.legend which identifies each marker
# I gave them greek labels to show that feature
# Again, alpha affects transparency 
# For plots I usually use 1 or something close to it, but I've showed you different options
# The order in which you plot determines which is plotted over which
plt.plot(x,y0,ls='',marker='s',ms=4,mfc='royalblue',mec='indigo',label=r'Data set $\alpha$',alpha=1)
plt.plot(x,y1,ls='',marker='o',ms=6,mfc='darkorange',mec='k',label=r'Data set $\beta$',alpha=0.8)
plt.plot(x,y2,ls='',marker='^',ms=8,mfc='forestgreen',label=r'Data set $\gamma$',alpha=0.6)

# This actually places the legend
# framealpha adjusts the transparency of the legend
# Set location to things like 'upper right', 'lower left', 'upper center', or just 'best' for automatic placement
# Fontsizes can be tricky here but whatever fits and is legible
# ncol sets the number of columns in the legend
plt.legend(loc="lower center", fontsize=10, ncol=3,framealpha=1)


# Save your figure
# I use .pdf for presentations and .eps when it needs to go into a LATEX document
# I usually keep this commented out until I'm ready to save it
#plt.savefig('/path/to/your/savefile')

# Display your plot to preview it
plt.show()
```

### Scatter Plots with Error Bars

```python
# I'll use the same data as the fake data set \alpha and \beta to show different error bar styles

#y-values for the first sample
yerr0 = np.random.uniform(low=1., high=2., size=(26,))

#x-values for the first sample
xerr1 = np.random.uniform(low=1.5, high=2.5, size=(26,))

#y-values for the first sample
yerr1 = np.random.uniform(low=1.5, high=2.5, size=(26,))
```

```python
# I'll exclude previous comments to only highlight the addition of error bars
# I've also changed the marker style to show you more options

a = plt.axes([0.18, 0.18, .64, .64])
plt.grid(True,ls='-.',alpha=.4)
plt.title(r'Meaningless Plots:'+'\n'+'Scatter with Errors',fontsize=18)
plt.xlabel(r'$X$ values',fontsize=16)
plt.ylabel(r'$Y$ values',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(-5,25)
plt.xlim(-4,53)

# I like the plt.errorbar function
# yerr and xerr are the various uncertainties, if one is not set it assumes it to be zero
# If you want edges around markers, set mfc and then color will be the edge and error bar color
# The default is no caps but if you want them set capsize
# ms still controlls marker size
plt.errorbar(x,y0,yerr=yerr0,ls='',color='k',capsize=3,mfc='indianred',ms=6,marker='>',label=r'Data set $\alpha$',alpha=0.9)
plt.errorbar(x,y1,yerr=yerr0,xerr=yerr0,ls='',color='orchid',marker='h',label=r'Data set $\beta$',alpha=1)

plt.legend(loc="lower right", fontsize=10, ncol=1,framealpha=1)
#plt.savefig('/path/to/your/savefile')
plt.show()
```

### Plotting Lines

```python
# Redefine fake data for smooth functions
# Actually I'm just using the functions that the previous data sets were based on

# x-values (smoother)
x_lin = np.linspace(-10,60,200)

#y-values for the function
y0_lin = 5.+0.3*x_lin

#y-values for the function
y1_lin = 0.005*x_lin**2.

#y-values for the function
y2_lin = 20.*np.exp(-x_lin/3.)
```

```python
a = plt.axes([0.18, 0.18, .64, .64])
plt.grid(True,ls='-.',alpha=.4)
plt.title(r'Meaningless Plots:'+'\n'+'Solid Lines',fontsize=18)
plt.xlabel(r'x values',fontsize=16)
plt.ylabel(r'y values',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(-2,25)
plt.xlim(-3,53)

# I also use plt.plot to make lines
# Without marker set to anything, it will do just that
# ls still sets the style, try different ones
# lw sets the width
# There really isnt much to this one
plt.plot(x_lin,y0_lin,ls='--',color='darkcyan',lw=2,label=r'Fit $\alpha$',alpha=0.9)
plt.plot(x_lin,y1_lin,ls='-',color='tan',lw=3,label=r'Fit $\beta$',alpha=0.9)
plt.plot(x_lin,y2_lin,ls=':',color='firebrick',lw=4,label=r'Fit $\gamma$',alpha=0.9)

plt.legend(loc="upper center", fontsize=10, ncol=1,framealpha=1)
#plt.savefig('/path/to/your/savefile')
plt.show()
```

### Histograms

```python
# Define a few fake distributions

# Lets define two different gaussian distributions
dist0 = np.random.normal(0, 0.1, 500)
count0, bin_edges0 = np.histogram(dist0,30,normed=False)
bin_centers0 = (bin_edges0[1:]+bin_edges0[:-1])/2

# This one will be taller and thinner, and shifted left
dist1 = np.random.normal(-0.2, 0.05, 500)
count1, bin_edges1 = np.histogram(dist1,30,normed=False)
bin_centers1 = (bin_edges1[1:]+bin_edges1[:-1])/2
```

```python
# Plotting just one distribution is easier
a = plt.axes([0.18, 0.18, .64, .64])

# Grid lines are up to you. They look good on some histograms, bad on others
plt.grid(True,ls='-.',alpha=.4)

# Usual plot features
plt.title(r'Meaningless Plots:'+'\n'+'Single Distribution',fontsize=18)
plt.xlabel(r'Bins',fontsize=16)
plt.ylabel(r'Counts',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(0,55)
plt.xlim(-0.4,0.4)

# For these, I use plt.bar 
# Like the markers, these can have edges or not, its up to you
# The third input is binwidth. I use bin_centers0[1]-bin_centers0[0] to ensure the bins are properly sized
plt.bar(bin_centers0,count0,bin_centers0[1]-bin_centers0[0],color='darkorchid',alpha=0.9,ec='k',label='Arbitrary Gaussian')

plt.legend(loc="best", fontsize=10, ncol=1,framealpha=1)
#plt.savefig('/path/to/your/savefile')
plt.show()
```

```python
# This is the same distribution but sampled and plotted differently
# In reality this is probably oversampled, but this is only for demonstration
count0_alt, bin_edges0_alt = np.histogram(dist0,80,normed=False)
bin_centers0_alt = (bin_edges0_alt[1:]+bin_edges0_alt[:-1])/2

# Usual plot features (I removed gridlines here)
a = plt.axes([0.18, 0.18, .64, .64])
plt.title(r'\textit{Meaningless Plots:}'+'\n'+r'\textit{Single Distribution}',fontsize=18)
plt.xlabel(r'\textit{Bins}',fontsize=16)
plt.ylabel(r'\textit{Counts}',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(0,23)
plt.xlim(-0.4,0.4)
plt.bar(bin_centers0_alt,count0_alt,bin_centers0_alt[1]-bin_centers0_alt[0],color='mediumseagreen',alpha=1.,label=r'\textit{Re-sampled}'+'\n'+r'\textit{Arbitrary Gaussian}')
plt.legend(loc="upper left", fontsize=10, ncol=1,framealpha=1)
#plt.savefig('/path/to/your/savefile')
plt.show()
```

```python
# Adjust the distributions for plt.step
# This trick adds an extra value of zero on the left and right of each histogram
# This is purely aesthetic so be careful you don't mess up your data
temp_bins0 = list(bin_centers0)
new_bins0 = [bin_centers0[0]-(bin_centers0[1]-bin_centers0[0])] + temp_bins0 + [bin_centers0[-1]+(bin_centers0[1]-bin_centers0[0])]

temp_count0 = list(count0)
new_count0 = [0] + temp_count0 + [0]

temp_bins1 = list(bin_centers1)
new_bins1 = [bin_centers1[0]-(bin_centers1[1]-bin_centers1[0])] + temp_bins1 + [bin_centers1[-1]+(bin_centers1[1]-bin_centers1[0])]

temp_count1 = list(count1)
new_count1 = [0] + temp_count1 + [0]
```

```python
# Here I'll plot two distributions on the same plot

# Usual plot features
a = plt.axes([0.18, 0.18, .64, .64])
plt.grid(True,ls='-.',alpha=.4)
plt.title(r'\textbf{Meaningless Plots:}'+'\n'+r'\textbf{Single Distribution}',fontsize=18)
plt.xlabel(r'\textbf{Bins}',fontsize=16)
plt.ylabel(r'\textbf{Counts}',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(0,51)
plt.xlim(-0.4,0.4)

# Rather than bars, I sometimes use plt.step if I have to plot multiple distributions
plt.step(new_bins0,new_count0,linestyle='-',label=r'Dist. $\mathcal{A}$',color='darkblue',linewidth=2)
plt.step(new_bins1,new_count1,linestyle='-',label=r'Dist. $\mathcal{B}$',color='goldenrod',linewidth=2)

plt.legend(loc="center right", fontsize=10, ncol=1,framealpha=1)
#plt.savefig('/path/to/your/savefile')
plt.show()
```

```python
# In all of these plots I've kept the same axes dimensions and gridlines
# Both can be changed according to your preference
# For example, plt.grid takes ls, alpha, and color so change those as you wish
# plt.grid(True,ls='[your choice]',alpha='[your choice]',color='[your choice]')

# Another useful feature is adding lines on a plot to mark locations
# Use plt.axvline for vertical lines and plt.axhline for horizontal
# It can be labeled as well to indicate what you're marking
# e.g., plt.axvline(100,ls='--',label=r'Special x-value') places a dashed vertical line at x = 100
# axhline works the same way, the value is the y-value at which to place a line
```

```python
# Matplotlib has a lot of colors that can be hard to remeber. This displays them all with names
# This code is from https://matplotlib.org/examples/color/named_colors.html
# Be careful when choosing colors, some look better than others

from __future__ import division
from matplotlib import colors as mcolors
colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]
n = len(sorted_names)
ncols = 4
nrows = n // ncols + 1
fig, ax = plt.subplots(figsize=(8, 5))
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols
for i, name in enumerate(sorted_names):
    col = i % ncols
    row = i // ncols
    y = Y - (row * h) - h
    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)
    ax.text(xi_text, y, name, fontsize=(h),horizontalalignment='left',verticalalignment='center')
    ax.hlines(y + h * 0.1, xi_line, xf_line,color=colors[name], linewidth=(h * 0.6))
ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()
fig.subplots_adjust(left=0, right=1,top=1, bottom=0,hspace=0, wspace=0)
plt.show()
```
