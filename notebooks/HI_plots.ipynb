```python
from astropy.table import QTable, Table

import numpy as np

import matplotlib.pyplot as plt
```

# Data

```python
data_filename = '../spirals/Pipe3D-master_file_vflag_BB_chi10_alpha10_smooth2p27.txt'

data = QTable.read(data_filename, format='ascii.ecsv')
```

```python
ALFALFA_filename = '../data/HI/manga_dr15_alfalfa.fits'
GBT_filename = '../data/HI/mangaHIall.fits'

ALFALFA = Table.read(ALFALFA_filename, format='fits')
GBT = Table.read(GBT_filename, format='fits')
```

# Distribution of WF50

```python
%matplotlib notebook

plt.figure()

#plt.hist(data['WF50'][np.logical_and(data['WF50'] < 0, data['WF50'] != -99)], bins=100)
plt.hist(ALFALFA['WF50'], bins=100, alpha=0.5)
plt.hist(GBT['WF50'], bins=100, alpha=0.5)

plt.xlabel('WF50 [km/s]')

plt.tight_layout();
```
