```python
from astropy.table import Table

import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
filename = '../spirals/Pipe3D-master_file_vflag_BB_minimize_chi10_smooth2p27_mapFit_N2O2_HIdr2_noWords_v4.txt'

data = Table.read(filename, format='ascii.commented_header')
```

```python
bad_fits = np.logical_or(data['M90_map'] == -99, data['M90_disk_map'] == -99)

good_fits = data[~bad_fits]
```

# Disk fit

```python
plt.figure()

plt.hist(good_fits['Sigma_disk_map'], bins=100)

plt.xlabel('$\Sigma_d$');
```

```python
gal_ID = '7443-12705'

plate, IFU = gal_ID.split('-')

galaxy_index = np.logical_and(good_fits['MaNGA_plate'] == int(plate), 
                              good_fits['MaNGA_IFU'] == int(IFU))

good_fits[['phi_map', 'ba_map', 'x0_map', 'y0_map', 'Sigma_disk_map', 'Rdisk_map']][galaxy_index]
```

```python
plt.figure()

plt.hist(good_fits['Rdisk_map'], bins=100)

plt.xlabel('$R_d$');
```
